## Task 2 

### Task 2.1: importing covid19 dataset
importing "Covid19_Confirmed_dataset.csv" from "./Dataset" folder. 


```python
data_corona=pd.read_csv("covid19_Confirmed_dataset.csv")
data_corona.head()
```

```python
data_corona.loc[data_corona['Country/Region']=='India']
```

### Task 2.2: Delete the useless columns

```python
data_corona.drop(columns=['Lat','Long'],inplace=True)
```

```python
data_corona.head()
```

### Task 2.3: Aggregating the rows by the country

```python
data=data_corona.groupby("Country/Region").sum()
```

```python
data
```

```python
data.shape
```

### Task 2.4: Visualizing data related to a country for example China
visualization always helps for better understanding of our data.

### Task3: Calculating a good measure

### task 3.1: caculating the first derivative of the curve

```python
data.loc['China'].diff().plot()
```

### task 3.2: find maxmimum infection rate for China

```python
data.loc['China'].diff().max()
```

```python
data.loc['United Kingdom'].diff().max()
```

```python
data.loc['India'].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries=list(data.index)
max_infections=[]
for c in countries:
    max_infections.append(data.loc[c].diff().max())
data['max_infections']=max_infections
```

```python
data.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_data=pd.DataFrame(data['max_infections'])
```

```python
corona_data.head()
```
